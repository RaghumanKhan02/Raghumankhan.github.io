```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv("C:/Users/91790/Downloads/archive (2)/Time_Spending_in_Social_Media_Dataset.csv")
```

```python
# Define income bins
income_bins = [10000, 12000, 14000, 16000, 18000, 20000]  # Adjust bins as needed
income_labels = ['10000-11999', '12000-13999', '14000-15999', '16000-17999', '18000-19999']
```

```python
# Create a new column for income bins
df['Income Bin'] = pd.cut(df['income'], bins=income_bins, labels=income_labels, right=False)
```

```python
# Group by age bins and calculate the sum of time spent
age_bins = [18, 30, 40, 50, 60, 65]  # You can adjust these bins as needed
age_labels = ['18-29', '30-39', '40-49', '50-59', '60+']
df['Age Bin'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)
age_time_spent = df.groupby('Age Bin')['time_spent'].sum().reset_index()
```

```python
# Print age-wise sum of time spent
print("Age-wise Sum of Time Spent:")
print(age_time_spent)
```

```python
# Visualize age bins and sum of time spent
plt.figure(figsize=(10, 6))
sns.barplot(data=age_time_spent, x='Age Bin', y='time_spent')
plt.title('Sum of Time Spent by Age Bin')
plt.xlabel('Age Bin')
plt.ylabel('Sum of Time Spent (hours)')
plt.show()
```

```python
# Group by income bins and calculate the sum of time spent
income_time_spent = df.groupby('Income Bin')['time_spent'].sum().reset_index()
```

```python
# Print income-wise sum of time spent
print("\nIncome-wise Sum of Time Spent:")
print(income_time_spent)
```

```python
# Visualize income bins and sum of time spent
plt.figure(figsize=(10, 6))
sns.barplot(data=income_time_spent, x='Income Bin', y='time_spent')
plt.title('Sum of Time Spent by Income Bin')
plt.xlabel('Income Bin')
plt.ylabel('Sum of Time Spent (hours)')
plt.show()
```

```python

# Group by gender, platform, interests, location, demographics, and profession
sum_time_spent = df.groupby(['gender', 'platform', 'interests', 'location', 'demographics', 'profession'])['time_spent'].sum().reset_index()
```

```python
# Print gender-wise, platform-wise, interests-wise, location-wise, demographics-wise, and profession-wise sum of time spent
print("\nGender-wise, Platform-wise, Interests-wise, Location-wise, Demographics-wise, and Profession-wise Sum of Time Spent:")
print(sum_time_spent)
```

```python
# Separate and visualize gender-wise, platform-wise, interests-wise, location-wise, demographics-wise, and profession-wise analysis
columns_to_plot = ['gender', 'platform', 'interests', 'location', 'demographics', 'profession']

for column in columns_to_plot:
    plt.figure(figsize=(12, 6))
    sns.barplot(data=sum_time_spent, x=column, y='time_spent')
    plt.title(f'Sum of Time Spent by {column.capitalize()}')
    plt.xlabel(f'{column.capitalize()}')
    plt.ylabel('Sum of Time Spent (hours)')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```
